>아나그램 탐지
- 아나그램(Anagram)은 한 문자열의 문자를 재배열해서 다른 뜻을 가지는 다른 단어로 바꾸는 것을 의미합니다.

- 두 개의 문자열이 주어졌을 때, 서로가 서로의 아나그램인지 아닌지의 여부를 탐지하는 함수를 만들어 보세요.

- elice 와 leice 는 아나그램입니다. True를 리턴해야 합니다.
- cat 과 cap 는 아나그램이 아닙니다. False 를 리턴해야 합니다.
- iamlordvoldemort 와 tommarvoloriddle 은 아나그램입니다. True를 리턴해야 합니다.
- 문자열의 모든 문자는 영어 소문자라고 가정합시다.

```python
def isAnagram(str1,str2):
    if len(str1) != len(str1):
        return False
    
    finding = list(str1)
    target = list(str2)
    
    for i in finding:
        if i in target:
            continue
        else:
            return False
    
    return True
    
    
print(isAnagram('cbp', 'cat')) 
```

엘리스 토끼는 단어장에 A와 B를 일렬로 적고 있었습니다. 그 모습을 본 도도새는 A는 A끼리 파란색으로, B는 B끼리 빨간색으로 글자 위에 아치형 곡선을 아래 그림과 같이 그렸습니다.


그런데 도도새는 ABAB처럼 파란 선과 빨간 선이 뿐만 아니라 두 선이 겹치는 것을 보기 싫어합니다. 그래서 선이 서로 겹치는 단어는 ‘도도새 취향이 아닌 단어’, 선이 겹치지 않는 단어는 ‘도도새 취향 단어’로 이름 붙였습니다.

단어가 주어졌을 때, ‘도도새 취향 단어’의 개수를 세는 프로그램을 작성하세요.

조건
하나의 문자는 동일한 글자인 문자와 단 하나의 곡선으로만 연결됩니다.
‘도도새 취향의 단어’가 되려면 연결되지 않은 문자가 없어야 합니다.
예를 들어 AAA는 문자 하나가 연결될 수 없으므로 ‘도도새 취향의 단어’가 아닙니다.

[입력]
첫째 줄에 단어의 수 N을 입력합니다.
(1≤N≤100)(1 ≤ N ≤ 100)
(1≤N≤100)
다음 N개 줄에는 A와 B로만 이루어진 단어를 한 줄에 하나씩 입력합니다.
단어의 길이는 2와 1,000 사이이며, 모든 단어 길이의 합은 10,000을 넘지 않습니다.

```python
def solution(string):
    stack = []
    for s in string:
        #1
        if not stack:
            stack.append(s)
            continue
        #2 stack의 top과 탐색할 문자열이 같으면 인접한 문자열
        if s == stack[-1]:
            stack.pop()
        else:
            stack.append(s)
        
    return stack
            



if __name__ == "__main__":
    n =int(input())
    count=0
    for _ in range(n):
        string = input()
        stack = solution(string)
        if len(stack) == 0:
            count+=1
    print(count)
            
```

- size: 스택에 들어있는 정수의 개수를 출력한다.
- empty: 스택이 비어있으면 1, 아니면 0을 출력한다.
- top: 스택의 가장 위에 있는 정수를 출력한다. 만약 스택에 들어있는 정수가 없는 경우에는 -1을 출력한다.

```python
class stackorder:
    def __init__(self):
        self._list = []
        
    def push(self,v):
        self._list.append(v)
        
    def pop(self):
        if self.empty() == -1:
            print(-1)
        else:
            print(self._list.pop())
    
    def size(self):
        print(len(self._list))
        return len(self._list)
    
    def empty(self):
        if self.size() == 0:
            print(-1)
            return -1
        else:
            print(0)
        
    def top(self):
        if self.size() == 0:
            print(-1)
        else:
            print(self._list[-1])
        

newstack = stackorder()

N = int(input())

for _ in range(N):
    order = list(input().split())
    if order[0] == "push":
        newstack.push(order[1])
    elif order[0] == "pop":
        newstack.pop()
    elif order[0] == "size":
        newstack.size()
    elif order[0] == "empty":
        newstack.empty()
    elif order[0] == "top":
        newstack.top()


```
