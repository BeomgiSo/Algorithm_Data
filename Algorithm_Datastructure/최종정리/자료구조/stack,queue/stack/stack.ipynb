> 스택
- `한쪽 끝에서만` 자료를 넣고 뺄 수 있는 자료구조
- 스택이 지언하는 `연산의 목록`
    - push : 스택에 자료를 넣는 연산
    - pop : 스택에서 자료를 뺴는 연산
    - top : 스택의 가장 위에 있는 자료를 반환하는 연산
    - empty : 스택이 비어있는지 여부를 반환하는 연산
- `LIFO(last in first out)`

- ex
    - push 1
    - 1
    - push 2
    - 1 2 
    - pop
    - 1
    - push 3
    - 1 3

- 나중에 들어간 자료가 먼저 출력되기 때문에 `Last in First Out(LIFO)`자료 구조이다.
- 파이썬의 리스트는 `append, pop등의 연산을 지원`한다.
- `리스트`를 이용하여 `스택`을 구현할 수 있다.

```python
# stack의 구현
class stack:
    def __init__(self):
        self.stack=[]
        
    def push(self,n):
        self.stack.append(n)
    
    def pop(self,n):
        if self.empty() == 1:
            return
        else:
            self.stack.pop()
            
    def size(self):
        return len(self.stack)
    
    def empty(self):
        if self.size() == 0:
            return 1
        else:
            return 0
        
    def top(self):
        if self.empty() == 1:
            return -1
        else:
            return self.stack[-1]


n = int(input())

mystack = stack()

for _ in range(n):
    order = [ int(v) for v in input().split()]
    if order[0] == 1:
        mystack.push(order[1])
    elif order[0] == 2:
        mystack.pop()
    elif order[0] == 3:
        print(mystack.size())
    elif order[0] == 4:
        print(mystack.empty())
    elif order[0] == 5:
        print(mystack.top())
    
               
```

```python
# 정답
class Stack:
    '''
    List를 이용하여 다음의 method들을 작성하세요.
    '''
    def __init__(self) :
        '''
        자료를 저장할 공간(배열) myStack을 만듭니다.
        '''
        self.myStack = []

    def push(self, n) :
        '''
        stack에 정수 n을 넣습니다.
        '''
        self.myStack.append(n)

    def pop(self) :
        '''
        stack에서 가장 위에 있는 정수를 제거합니다. 만약 stack에 아무 원소가 없다면 아무 일도 하지 않습니다.
        '''
        if self.empty() == 1:
            return
        else:
            self.myStack.pop()

    def size(self) :
        '''
        stack에 들어 있는 정수의 개수를 return 합니다.
        '''
        return len(self.myStack)

    def empty(self) :
        '''
        stack이 비어있다면 1, 아니면 0을 return 합니다.
        '''
        if self.size()==0:
            return 1
        else:
            return 0

    def top(self) :
        '''
        stack의 가장 위에 있는 정수를 return 합니다. 만약 stack에 들어있는 값이 없을 경우에는 -1을 return 합니다.
        '''
        if self.empty() == 1 :
            return -1
        else : 
            return self.myStack[-1]
def find_order(p) :
    '''
    괄호 p가 주어질 때, 각 괄호가 몇 번째로 계산되어야 하는 괄호인지를 list로 반환합니다.

    예를 들어, p='(()())' 일 경우, [3, 1, 1, 2, 2, 3] 을 반환합니다.
    '''
    s = Stack()
    
    result = [0] * len(p)
    
    cnt = 1
    
    for i in range(len(p)):
        if p[i] == "(":
            s.push(i)
        else : 
            index = s.top()
            s.pop()
            
            result[index] = cnt
            result[i] = cnt
            
            cnt = cnt + 1

    return result
```

```python
class Queue:
    def __init__(self):
        self.queue = []
    
    def push(self,n):
        self.queue.append(n)
    
    def pop(self):
        if self.empty() == 1:
            return
        del self.queue[0]
    
    def front(self):
        if self.empty() == 1:
            return -1
        return self.queue[0]
    
    def back(self):
        if self.empty() == 1:
            return -1
        
        return self.queue[-1]
```

- `언더플로(Underflow)` 특정한 자료구조에 데이터가 전혀 들어 있지 않은 상태에서 삭제 연산을 수행하면 데이터가 전혀 없는 상태인 언더플로우 발생

# 연습문제

- 괄호 쌍이 주어질 때, 각 괄호가 몇 번째로 계산되어야 하는 괄호인지를 출력하는 find_order(p)를 작성하세요.

- 입력
첫 번째 줄에 길이가 n인 괄호 문자열 p가 주어집니다. 2≤n≤200002 \le n \le 200002≤n≤20000

주어지는 문자열 p는 항상 올바른 괄호 문자열입니다. 즉, ‘)(‘이나 ‘())’ 과 같은 입력은 주어지지 않습니다.

- 출력
각 괄호가 몇 번째로 계산되어야 하는 괄호인지 공백으로 구분한 결과를 리스트로 담아 반환합니다.

```python

a = input()
def find_order(p):
    result = [0]*len(p)
    
    cnt = 1
    
    stack = []
    
    for i in range(len(p)):
        if p[i] == "(":
            stack.append(i)
        else:
            index = stack[-1]
            stack.pop()
            
            result[index] = cnt
            result[i] = cnt
            
            cnt = cnt+1
            
    return result

print(find_order(a))
```
