```python
# greedy 문제
# 이코테 #1
# N = int(input())
# data = list(map(int,input()))

# 조건 N 명이 있고 X인 모험가는 반드시 X명 이상으로 구성한 모험가의 그룹을 만들어야 여행으 떠날 수 있다.
# 이때 여행을 떠날 수 있는 그룹수의 최댓값을 출력하자.
N = 5
data = [2,3,1,2,2]
data.sort()
# 최대의 갯수를 뽑으려면? => 낮은애들끼리 합 5 이상인지 확인
check_num = []
cnt = 1
for i in data:
    if sum(check_num) < N:
        check_num.append(i)
    else:
        cnt +=1
        check_num = [i]

print(cnt)
        
```

```python
# 곱하기 더하기에 대하여
# 이코테 #2
# 틀림 다시 생각해보기
N = list(map(int,input()))
result = 1
for i in N:
    if i*result == 0:
        result +=i
    else:
        result *=i
    
print(result)
# 02984
# 2023
```
