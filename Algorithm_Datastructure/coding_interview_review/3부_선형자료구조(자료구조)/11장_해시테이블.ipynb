# 11장 해시 테이플
- 해시 테이블 또는 해시 맵은 키를 값에 매핑할 수 있는 구조인, 연관 배열 추상자료형(ADT)을 구현하는 자료구조다.
- 해시테이블의 가장 큰 특징은 대부분 연산이 분할 상환 분석에 따른 시간 복잡도가 O(1)이라는 점이다.
- 따라서, 데이터 양에 관계 없이 빠른 성능을 기대할 수 있다는 장점이 있다.

> 해시
- 해시 함수란 임의 크기 데이터를 고정 크기 값으로 매핑하는 데 사용할 수 있는 함수를 말한다.
- 해시 테이블의 핵심은 해시 함수다.

- ex) 특정함수를 통과하면 2바이트의 고정 크기 값으로 매핑된다.
    - ABC -> A1
    - 132BC -> CB
    - AF32B -> D5

- 해시 테이블을 인덱싱하기 위해 이처럼 해시 함수를 사용하는 것을 해싱(hashing)이라 하며, 해싱은 정보를 가능한 빠르게 저장하고 검색하기 위해 사용하는 중요한 기법이다.
- 해싱은 최적의 검색이 필요한 분야에 사용되며, 심볼 테이블(일반적으로 해시 테이블로 구현된다.)등의 자료구조를 구현하기에도 적합하다.
- 이외에도 해시 함수는 체크섬(Checksum), 손실 압축, 무작위화 함수(Randomization Function), 암호 등과도 관련이 깊으며 떄로는 서로 혼용되기도한다.
- 그러나 어느 정도는 개념이 겹치긴하지만, 서로 용도와 요구사항이 다른만큼 다르게 설계되고 최적화 된다.

- 성능 좋은 해시 함수의 특징
    - 해시 함수 값 충돌의 최소화
    - 쉽고 빠른 연산
    - 해시 테이블 전체에 해시 값이 균일하게 분포
    - 사용할 키의 모든 정보를 이용하여 해싱
    - 해시 테이블 사용 효율이 높을 것

- 생일 문제(Birthday Problem)
- 충동의 예시. 생일의 가짓수는 365개이므로 여러사람이 모였을 때 생일은 같은 2명이 존재할 확률을 얼핏 생각하면 비둘기 집 원리에 따라 366명 이상이 모여야 일어나는 일 같다.
- 그러나 실제로는 23명만 모여도 50%가 높고, 57명만 넘어도 그떄부터 99%를 넘어선다.



```python
import random

TRIALS = 100000 # 10만번 실험
same_birthdays = 0 # 생일이 같은 실험의 수

# 10만 번 실험의 진행
for _ in range(TRIALS):
    birthdays = []
    # 23명이 모였을때
    for i in range(23):
        brithday = random.randint(1,365)
        if brithday in birthdays:
            same_birthdays +=1
            break
    birthdays.append(brithday)

    # 전체 10만번 실험 중 생일이 같은 실험의 확률
    print(f'{same_birthdays / TRIALS * 100}%')
```

> 비둘기집 원리
- 비둘기집 원리란, n개 아이템을 m개 컨테이너에 넣을 때, n>m이라면 적어도 하나의 컨테이너에는 반드시 2개이상의 아이템이 들어 있다는 원리
- 여러번 충돌한다는 것은 그만큼 추가 연산을 필요로 하기 떄문에 가급적 충돌은 최소화하는 것이 좋다.

> 로드 팩터(Load Factor)
- 로드 팰터란 해시 테이블에 저장된 개수 n을 버킷의 개수 k로 나눈 것이다

$$load factor = \frac{n}{k}$$

- 로드 팩터 비율에 따라서 해시 함수를 재작성해야 될지 또는 해시 테이블의 크기르 조정해야할지 결정해야한다.
- 그리고, 이 값은 해시 함수가 키들을 잘 분산해 주는지 말하는 효율성 측정에도 사용된다.
- 자바 10 에서는 해시맵의 디폴트 로드 팩터를 0.75를 정했으며 `시간과 공간 비용의 적절한 절충안`이라고 한다.
- 일반적으로 로드 팩터가 증가할 수록 해시 테이블의 성능은 점점 감소하게 되며, 자바10의 경우에는 0.75를 넘어설 경우 동적 배열처럼 해시 테이블의 공간을 재할당한다.

> 해시 함수
- 해시 테이블을 인덱싱하기 위해 해시 함수를 사용하는 것을 해싱(hashing)이라고 한다.
- 해싱에는 다양한 알고리즘이 있으며, 최상의 분포를 제공하는 방법은 데이터에 따라 제각각이다.
- 가장 기본적이고 널리 쓰이는 정수형 해싱 기법인 모듈로 연산을 이용한 나눈셈 방식(Modulo-Division Method)하나만 살펴본다.

$$h(x) = x \ mod \ m$$

    - 여기서 h(x)는 입력값 x의 해시 함수를 통해 생성된 결과다. m은 해시 테이블의 크기로, 일반적으로 2의 멱수에 가깝지 않은 소수를 택하는 것이 좋다.
    - h(x)는 이 값의 모듈로 연산의 결과로, 매우 단순한 방법이지만 실무에서는 이미 많은 키 세트가 충분히 랜덤한 상태고, 키 세트가 어떤 큰 소수에 의해 순환 구조가 될 확률이 낮기 떄문에 실제로는 잘 동작한다.

## 29번 문제 보석과 돌
- J는 보석이고,S는 갖고 있는 돌이다. S에는 보서기 몇개나 있을까? / 대소문자는 구분한다
- ex J='aA', S="aAAbbb"
- 3

```python

J = 'aA'
S="aAAbbb"

# 전형적인 freqs라는 해시 테이블
freq = {}

for char in S:
    if char not in freq:
        freq[char] = 1
    else:
        freq[char] +=1

count = 0
for char in J:
    if char in freq:
        count+= freq[char]

print(count)

# defaultdict을 이용한 비교생략
# defaultdict을 사용하면 존재하지 않은 키에 대하 디폴트를 리턴해주는 풀이를 가끔식 사용한다
# 키가 존재하느닞 여부를 매번 판별할 필요가 없기 때문에 바로 계싼할 수 있고 코드를 정리할 수 있다.
import collections
freqs = collections.defaultdict(int)
count = 0

# 비교없이 돌(S) 빈도 수 계산
for char in S:
    freqs[char] += 1

# 비교없이 보석(J)빈도 수 합산
for char in J:
    count += freqs[char]

print(count)


```

## 상위 K 빈도요소

```python
nums = [1,1,1,2,2,3]
k = 2

diction = collections.defaultdict(int)

for i in nums:
    diction[i] +=1

solution = []
diction = dict(diction)
for key,value in diction.items():
    if value>=k:
        solution.append(key)

print(solution)
```
