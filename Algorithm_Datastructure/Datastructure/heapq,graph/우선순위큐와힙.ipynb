> 1. 우선순위 큐의 구현 방법과 힙
- 우선순위 큐란?
- 우선순위가 높은 `원소`가 `먼저 출력`되는 추상적 자료형
- 출력 연산시 가장 `우선순위가 높은` 원소를 출력한다.

- 우선순위 큐를 배열로 단순하게 구현해보자
- 51234 -> 출력 5 4 3 2 1
- 입력 : O(1) // 출력 : O(N)

- 우선순위가 가장 높은 원소를 찾는 과정과, 그 원소를 제거하는 것이 비효울적이다.
- 더좋은 방법이 없을까?

>  힙
- `최댓값` 또는 `최솟값`을 빠르게 찾기 위해 고안된 `완전이진트리`
- 최대 힙(Max Heap)
    - 부모노드는 항상 자식노드보다 `큰` 값을 갖고 있다.
- 최소 힙(Min Heap)
    - 부모 노드는 항상 자식 노드보다 `작은` 값을 갖고 있다.

```python
import heapq
```
- 파이썬에서는 heapq 모듈로 `최소힙`을 사용할 수 있다.
- 최대 힙이 필요한 경우에는 값을 저장할 때 -1을 곱한 값을 저장하면된다.
- `-1 곱함`으로써 최대값과 최소값이 반전된다는 점을 이용하면 된다.
- 단 이방법은 힙이 저장하는 값이 `수(Number)`일 때만 가능하다.

- 힙은 `완전 이진트리`의 특성을 유지해야한다.

> 힙 : 자료 입력하기
- 최악의 경우는 `새로운 최솟값`이 입력되는 경우이고, `루트 노드`까지 거슬러 올라가게 된다.
